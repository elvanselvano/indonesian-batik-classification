```python
import shutil, os, time, uuid, sys, random
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from tqdm.auto import tqdm
from google.colab import drive
USER_AGENT = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36'
BUFFER_CHOICE_LIST = np.arange(0.1, 3, 0.3).tolist()
```

```python
def define_webdriver():
  chrome_options = webdriver.ChromeOptions()
  chrome_options.add_argument('--headless')
  chrome_options.add_argument('--no-sandbox')
  chrome_options.add_argument('--disable-dev-shm-usage')
  chrome_options.add_argument(f'user-agent={USER_AGENT}')
  chrome_options.add_argument("--window-size=1920,1080")
  return webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
def buffer():
  time.sleep(random.choice(BUFFER_CHOICE_LIST))
```

```python
def check_word_presence(strings, word, splitter = '\n'):
  return sum([i.lower().split(' ').count(word.lower()) for i in strings.split(splitter)]) != 0
```

```python
def init_dirs(dirs):
  for i in dirs:
    try:
      os.makedirs(i)
    except:
      shutil.rmtree(i)
      os.makedirs(i)
```

```python
def scrap_zalora(baseURL, pageRange = 10, keyword = True, dir = '.', extension = 'jpg'):
  with define_webdriver() as driver:
    driver.maximize_window()
    for i in range(pageRange):
      buffer()
      print(f'page: {i + 1}/{pageRange}')
      driver.get(baseURL + str(i+1))
      try:
        element = driver.find_element_by_class_name('b-catalogList__wrapper')
        lists = element.find_elements_by_tag_name('li')
      except:
        continue
      for list_item in tqdm(lists):
        if keyword or check_word_presence(list_item.text, keyword):
          try:
            img_wrapper = list_item.find_elements_by_tag_name('span')[1]
            hover = ActionChains(driver).move_to_element(img_wrapper)
            hover.perform()
            img_wrapper.screenshot(f'{dir}/{uuid.uuid4()}.png')
          except:
            continue
```

```python
def zip_dirs(dirs, file_prefix = '', file_sufix=''):
  zip_list = []

  for i in tqdm(dirs):
    zip_list.append(shutil.make_archive(file_prefix + i + file_sufix, 'zip', i))
  return zip_list
```

```python
def move_zips(files, dir):
  for i in tqdm(files):
    shutil.move(i, dir)
```
