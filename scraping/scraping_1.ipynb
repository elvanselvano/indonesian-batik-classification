```python
import os
import shutil
import requests
from google.colab import drive
from bs4 import BeautifulSoup
import re
import time
```

```python
def init():
  drive.mount('/content/gdrive')
  
  try:
    shutil.rmtree('motives')
    os.makedirs('motives')
  except:
    os.makedirs('motives')

  try:
    shutil.rmtree('batik_clothes')
    os.makedirs('batik_clothes')
  except:
    os.makedirs('batik_clothes')
```

```python
MOTIVES_COUNT = 0
BATIK_COUNT = 0
MOTIVES_DIR = 'motives'
BATIK_DIR = 'batik_clothes'
init()
```

```python
def getScrapHtml(url):
  html = requests.get(url)
  html = BeautifulSoup(html.content)
  return html
```

```python
def saveImgs(img_links, img_count, dir, extension = 'jpg'):
  for j in img_links:
    with open(f'{dir}/{img_count}.{extension}', 'wb') as file:
      file.write(requests.get(j).content)
      img_count += 1
      time.sleep(0.1)
  return img_count
```

```python
def getAllImgLink(html, attr = 'src'):
  return [i.get(attr) for i in html.find_all('img')]
```

```python
url = 'https://fasnina.com/motif-batik/#Model_Batik_Nusantara'
html = getScrapHtml(url)
```

```python
html = html.find('div', id='content')
figures = html.find_all('figure')
figures
```

```python
links = []
for i in figures:
  links.append(getAllImgLink(i)[1])
links
```

```python
motives_link = [i for i in links if re.search('motif', i)]
batik_link = [i for i in links if not re.search('motif', i)]
```

```python
MOTIVES_COUNT = saveImgs(motives_link, MOTIVES_COUNT, MOTIVES_DIR)
BATIK_COUNT = saveImgs(batik_link, BATIK_COUNT, BATIK_DIR)
```

```python
url = 'https://fasnina.com/batik-cirebon/'
html = getScrapHtml(url)
html = html.find('div', id='content')
figures = html.find_all('figure')
links = []
for i in figures:
  links.append(getAllImgLink(i)[1])
links
```

```python
MOTIVES_COUNT = saveImgs(links[:15], MOTIVES_COUNT, MOTIVES_DIR)
BATIK_COUNT = saveImgs(links[15:], BATIK_COUNT, BATIK_DIR)
```

```python
url = 'https://blog.mariberkarya.com/en/51-typical-motif-batik-in-yogyakarta-traditional-motif/'
html = getScrapHtml(url)
html = html.find('div', class_='entry-content')
figures = html.find_all('figure')
links = []
for i in figures:
  links.append(getAllImgLink(i)[1])
links
```

```python
MOTIVES_COUNT = saveImgs(links[1:], MOTIVES_COUNT, MOTIVES_DIR)
```
